# Spectrogram Chirp

$f (t) = cos(2\pi t \omega(t))$
where
$\omega(t) = \omega_0 + (\omega_1 âˆ’ \omega_0)t^2/3t^2_1$ .

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams.update({'font.size': 18})

dt = 0.001
t = np.arange(0,2,dt)
f0 = 50
f1 = 250
t1 = 2
x = np.cos(2*np.pi*t*(f0 + (f1-f0)*np.power(t,2)/(3*t1**2)))

plt.specgram(x, NFFT=128, Fs=1/dt, noverlap=120,cmap='jet')
plt.colorbar()
plt.show()
```

# HAAR

```python
def psit(a=1,b=0):
    t = np.arange(-0.25,1.25,0.001)
    tab = (t-b)/a
    #psi = np.zeros_like(x)
    psi = np.where((tab<0.5)&(tab>=0),1,0)
    psi = np.where((tab<1)&(tab>=0.5),-1,psi)
    psiab = psi/np.sqrt(a)
    #plt.plot(t,psiab)
    return psiab
psit(0.5)
ti = np.arange(-0.25,1.25,0.001)
plt.plot(ti,psit(),label = 'psi(1,0)', LineWidth=2)
plt.plot(ti,psit(0.5),'--',label = 'psi(0.5,0)')
plt.plot(ti,psit(0.5,0.5),'-o',label = 'psi(0.5,0.5)')
plt.legend()
plt.show()
```
